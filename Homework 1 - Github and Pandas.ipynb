## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
from numpy import *

# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations,learning_rate = 0.0001):
    points = genfromtxt("/Users/chi/Desktop/APMAE4990-/data/data.csv", delimiter=",")
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))     
          
    print("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
               
    print("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

run(100)
````

````python
points = genfromtxt("/Users/chi/Desktop/APMAE4990-/data/data.csv", delimiter=",")
learning_rates = np.logspace(-4.5,-3.5,100)
errors = []
initial_b = 0
initial_m = 0
num_iterations = 5
for l in learning_rates:
    learning_rate = l
    [b,m] = gradient_descent_runner(points,initial_b,initial_m,learning_rate,num_iterations)
    errors.append(compute_error_for_line_given_points(b , m ,points))

   
plt.plot(learning_rates,errors,)
plt.xlabel("Learning Rates")
plt.ylabel("Errors")

````

By modifying the learning rates of gradient descent method and looking at the U-shape plot(LearningRates/Errors), we can interpret that there are three stages of the convergence/divergence problem: faster/converge - fastest/converge - slow/diverge. Three results are blow:
![image](https://raw.github.com/chima222/python-introduction-chima222/image/Screen Shot 2018-02-26 at 5.12.32 PM.png)

From the gradient descent formula: ![image](https://raw.github.com/chima222/python-introduction-chima222/image/img8.png) 

Here, Alpha is the learning rate and determines the time steps that gradient descent mathod uses to minimize(not a global min necessarly) a given convex function theta. So, alpha is a weighting mechanism we use to determine how much emphasis we want to put on the gradient of the cost function. When the learning rate is too small, it takes a long time or more iterations for the convergence to happen: Case 1, learning rate is 0.000001 and it takes 1000 iterations to get an ideal convergence. If the learning rate is too large, then the result is very likely to diverge since the large step keeps the method from hit the minimium: Case 2, learning rate is 0.005 and it goes to diverge. When the ideal learning rate is choosen, it will converge in short period of time(low iterations): Case 2, it only takes 100 iterations for the errors to approach the abosult min errors.
