## How could you have improved performance?

*Hint:* Try plotting the data in three dimensions along with the hyperplane solution, and see if you can infer
a new variable which will help, or try a nonlinear/non-parametric model

````python
# Type your answer here. 
model = sm.ols(formula='sales ~ TV + radio', data = df)
fit = model.fit()

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

x_surf = np.arange(0, 320, 20)              
y_surf = np.arange(0, 60, 4)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

exog = pd.DataFrame({'TV': x_surf.ravel(), 'radio': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.values.reshape(x_surf.shape),cmap=cm.coolwarm,alpha = 0.4)
ax.scatter(df['TV'], df['radio'], df['sales'],
           c='red',)
ax.set_xlabel("TV")
ax.set_ylabel("radio")
ax.set_zlabel("sales")
plt.show()
````

Like the hyperplane shows, most of the points are in the plane but there are some outliers and indicates a nonlinear trend. We would either add an interation variable or polynomial variable or both to improve the performance.  

````python
#Add new variable
X["TV/radio"]=X["TV"]*X["radio"]
X3 = X.drop(["newspaper"],1)
#Standardize data
scaler = preprocessing.StandardScaler()
X3 = X3.as_matrix().astype(np.float)
X3 = scaler.fit_transform(X3)

size = float(len(X))

X3_train, X3_test, y_train, y_test = train_test_split(
     X3, y, test_size=0.2, random_state=42)


regr3 = LinearRegression()

regr3.fit(X3_train, y_train)

print('Coefficients: \n', regr3.coef_)
print("Residual sum of squares: %.2f"
       % np.mean((regr3.predict(X3_test) - y_test) ** 2))
print('Variance score: %.2f' % regr3.score(X3_test, y_test))
````

````python
plt.scatter(regr3.predict(X3_test),y_test)
plt.plot(y_test,y_test,color = "r")
plt.show()
````

````python
model = sm.ols(formula='sales ~ TV*radio', data = df)
fit = model.fit()

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

x_surf = np.arange(0, 320, 20)              
y_surf = np.arange(0, 60, 4)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

exog = pd.DataFrame({'TV': x_surf.ravel(), 'radio': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.values.reshape(x_surf.shape),cmap=cm.coolwarm,alpha = 0.4)
ax.scatter(df['TV'], df['radio'], df['sales'],
           c='red',)
ax.set_xlabel("TV")
ax.set_ylabel("radio")
ax.set_zlabel("sales")
plt.show()
````

## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
from numpy import *

# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations,learning_rate = 0.0001):
    points = genfromtxt("https://raw.githubusercontent.com/Columbia-Intro-Data-Science/APMAE4990-/master/data/data.csv", delimiter=",")
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print ("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

run(100)
````

````python
points = genfromtxt("https://raw.githubusercontent.com/Columbia-Intro-Data-Science/APMAE4990-/master/data/data.csv", delimiter=",")
learning_rates = logspace(-4.0,-3.5,100)
initial_b = 0
initial_m = 0
errors=[]
for l in learning_rates:
    [b,m] = gradient_descent_runner(points,initial_b,initial_m,l,10)
    errors.append(compute_error_for_line_given_points(b,m,points))
plt.plot(learning_rates,errors)
plt.xlabel("Learning_rate")
plt.ylabel("Errors")
plt.show()
````

From the mathemetical equation we can see that:


The learning rate(alpha) is the time step that the gradient descent method uses to minimize a given convex function(not necessarly a globel min). Alpha is also the emphasis that we want to put on the gradient of the cost function. If alpha is small, the difference between theta is small, hence the result will be more accurate but the calculation will be time consuming; if alpha is too big, the difference will increase and it is very likely we will swing back and forth around the minimum and never hit it; if the alpha is a proper value, we will be able to hit minimum accurately and fast. 


In this specific example, like the graph shows, when we modifying the learning rate with fixed iterations, the result shows three stages: starting converge: steps smaller than 0.0001(slow speed) --> fastest converge: steps from 0.0001 to 0.0003 --> diverge: steps larger than 0.00003.
